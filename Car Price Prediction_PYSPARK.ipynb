```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Price Prediction').getOrCreate()
spark
```

```python
data = spark.read.csv("C:\\AI(ml&dl)\\jupiter(machine learning)\\Projetcs(ML)\\16_CarPricePrediction_RANDOMFOREST\\dataset.csv",header = True,inferSchema = True)
data.show(5)
```

```python
data.printSchema()
```

```python
data = data.drop('Car_ID')
```

```python
from pyspark.ml.feature import StringIndexer
indexer = StringIndexer(inputCols = ['CarName','fueltype', 'aspiration', 'doornumber', 'carbody','drivewheel', 'enginelocation',
                        
        'enginetype','cylindernumber','fuelsystem'],
                        outputCols=['{}_indexed'.format(c) for c in ['CarName','fueltype', 'aspiration', 'doornumber', 'carbody','drivewheel', 'enginelocation',
                        
        'enginetype','cylindernumber','fuelsystem']])
data = indexer.fit(data).transform(data)
data.toPandas()
                        
                    
                        
  
```

```python
data = data.drop('CarName','fueltype', 'aspiration', 'doornumber', 'carbody','drivewheel', 'enginelocation',
                        
        'enginetype','cylindernumber','fuelsystem')
```

```python
data.show()
```

```python
#checking the nullvalues

from pyspark.sql.functions import isnan,count,col,when
data.select([count(when(col(c).isNull(),c)).alias(c) for c in data.columns]).show(5)
```

```python
#checking the correlation 

data.toPandas().corr()
```

```python
new_data = data.drop('price')
new_data.toPandas()
```

```python
cols = new_data.columns
cols
```

```python
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols = cols,outputCol = 'Independent Features')
data = assembler.transform(data)
data.show(5)
```

```python
finilized_data = data.select('Independent Features','price')
finilized_data.show(10)
```

```python
#split train test

train_data,test_data = finilized_data.randomSplit([0.75,0.25])
train_data.count(),test_data.count()
```

```python
#standarized
from pyspark.ml.feature import StandardScaler
SC = StandardScaler(inputCol = 'Independent Features',outputCol = 'ScaledFeatures')
train_data = SC.fit(train_data).transform(train_data)
test_data = SC.fit(test_data).transform(test_data)
```

```python
train_data.show(5)
```

```python
test_data.show(5)
```

```python
#Linear Regression model
from pyspark.ml.regression import LinearRegression
LR = LinearRegression(labelCol = 'price',featuresCol = 'Independent Features')
LR = LR.fit(train_data)

```

```python
res = LR.evaluate(test_data)
res.predictions.show(10)
```

```python
print(f"""Logistic Regression Results Report
=======================================
r2:\t {res.r2*100}
RMSE:\t {res.rootMeanSquaredError}
MSE:\t {res.meanSquaredError}
Adj R2:\t {res.r2adj*100}
=======================================""")
```
